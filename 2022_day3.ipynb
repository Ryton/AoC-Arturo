```python
def split_in_threes(i=None):
    list_w_lines = i.split('\n')
    # print(list_w_lines)
    nlines = len(list_w_lines)
    badge = []

    for a in range(int(nlines/3)):
        linenr = a*3
        # print(linenr)
        found = False
        for l in list_w_lines[linenr]:
            if l in list_w_lines[linenr+1] and l in list_w_lines[linenr+2]:
                if not found:
                    badge.append(l)
                    found = True
    return badge


badge = split_in_threes(iA)
# print(double_letter)
print(len(badge))
print('set:', [prio_dict[o] for o in set(badge)])
print('sum_of_set:', sum([prio_dict[o] for o in set(badge)]))
print('sum_of_list:', sum([prio_dict[o] for o in badge]))
answ_b = sum([prio_dict[o] for o in set(badge)])
```

```python
def test_yield(init = 5):
    memory = init   
    add = 2
    while True:
        memory += add
        message = yield memory + add
        if message != None:
            add = message[0]
        else:
            add = 0

f_using_yield = test_yield(2.5)
f_using_yield.send(None)
print(next(f_using_yield))
f_using_yield.send([1, 0, 0])
print(next(f_using_yield))
f_using_yield.send([2, 0, 0])
print(next(f_using_yield))
f_using_yield.send([3, 0, 0])
print(next(f_using_yield))

```
