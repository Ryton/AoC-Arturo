```python
def paper_unfold(paper_x,paper_y, x_y_tuple):
    if not x_y_tuple[0]==0:
        paper_x = x_y_tuple[0]*2 + 1
    elif not x_y_tuple[1]==0:
        paper_y =  x_y_tuple[1]*2 + 1
        
    return paper_x,paper_y

paper_x,paper_y = 39,5

folds = [[0,6],[0,13],[0,27],[40,0],[0,55],[81,0],[0,111],[163,0],[0,223],[327,0],[0,447],[655,0]]
for fold in folds:
    
    print(fold)
    paper_x,paper_y=paper_unfold(paper_x,paper_y, fold)
    print(paper_x,paper_y)
    
print(paper_x,paper_y)
```

```python

def fold_one(paper,fold_at_xy):
    newpaper = paper
    if not fold_at_xy[0]==0: #fold x-direction
        for x in range(fold_at_xy[0]):
            #print(x,2*fold_at_xy[0]-x)
            newpaper[x,:]+= paper[2*(fold_at_xy[0])-x,:]
        return newpaper[0:fold_at_xy[0],:]
    elif not fold_at_xy[1]==0:
        for y in range(fold_at_xy[1]):
            newpaper[:,y] += paper[:,2*fold_at_xy[1]-y]
        return newpaper[:,0:fold_at_xy[1]]
```
