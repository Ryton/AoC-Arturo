```python
df =pd.read_csv('input10.txt',header=None)
valuelist = df.values #full

print(len(valuelist),'rows')
n=15
digitn=5
print(len(valuelist[n,0]), ' digits/chars on line ',n)
print('excerpt:',valuelist[n,0][1:digitn])

```

```python
puzzle_input = valuelist[:,0]
print(puzzle_input)
```

```python
score =0
score_uncompleted=[]
for stringlist in puzzle_input:
    for iteration in range(100):
        stringlist= stringlist.replace('()','')
        stringlist= stringlist.replace('[]','')
        stringlist= stringlist.replace('{}','')
        stringlist= stringlist.replace('<>','')
    #print(stringlist)
    
    
    #): 3 points.
    #]: 57 points.
    #}: 1197 points.
    #>: 25137 points.
    
    b1 = stringlist.find(')')
    b2 = stringlist.find(']')
    b3 = stringlist.find('}')
    b4 = stringlist.find('>')
    
    if b1==-1:
        b1 = 999
    if b2==-1:
        b2 = 999
    if b3==-1:
        b3 = 999
    if b4==-1:
        b4 = 999
                
    first_wrong = min(b1,b2,b3,b4)
    #print(b1,b2,b3,b4,first_wrong)
    if first_wrong == 999:
        #not wrong, but uncompleted_close!
        print(stringlist)
        line_score=0
        for char in stringlist[::-1]: #todo: reverse order
            print(char)
            if char=='(':
                line_score=line_score*5+1
                print(char, line_score)
            elif char=='[':
                line_score=line_score*5+2
                print(char, line_score)
            elif char=='{':
                line_score=line_score*5+3
                print(char, line_score)
            elif char=='<':
                line_score=line_score*5+4
                print(char, line_score)
            #): 1 point.
            #]: 2 points.
            #}: 3 points.
            #>: 4 points.

        score_uncompleted.append(line_score)
        
    elif  first_wrong == b1:
        score += 3   
    elif  first_wrong == b2:
        score += 57
    elif  first_wrong == b3:
        score += 1197
    elif  first_wrong == b4:
        score += 25137
print('answ a:',score)
print('answ b:',np.median(np.array(score_uncompleted)))
```
