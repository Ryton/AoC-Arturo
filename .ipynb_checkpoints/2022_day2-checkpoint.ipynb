```python
# test and solve part A
#iA,iE, 
#wins_from_x {'B':'A','C':'B','A':'C'}
wins_from_x ={'Y':'A','Z':'B','X':'C'}
draw_from_x ={'Y':'B','Z':'C','X':'A'}
lose_from_x ={'Y':'C','Z':'A','X':'B'}

transform_a_to_x = {'B':'Y','C':'Z','A':'X'}
transform_x_to_a = draw_from_x

dict_scoring_shape = {'X':1,'Y':2,'Z':3}
dict_scoring_match = {'w':6,'d':3,'l':0}

#for Rock, B for Paper, and C for Scissors

class preprocess():
    def __init__(self,i = None):
        self.input = i
        
        pass
    def fun(self,params=None):
        print('hello')
        pass
def preprocess_to_tuples(i = None):
    tuple_list = []
    for pair in i.split('\n'):
        letters= pair.split(' ')
        tuple_list.append((letters[1],letters[0])) 
    return(tuple_list)
        
    
def scoring(tuple_list=None):

    score = 0
    for letterpair in tuple_list:
        if wins_from_x[letterpair[0]]==letterpair[1]:
            score +=dict_scoring_match['w'] +dict_scoring_shape[letterpair[0]]
        elif draw_from_x[letterpair[0]]==letterpair[1]:
            score +=dict_scoring_match['d']+dict_scoring_shape[letterpair[0]]
        elif lose_from_x[letterpair[0]]==letterpair[1]:
            score +=dict_scoring_match['l']+dict_scoring_shape[letterpair[0]]
    return score

def change_action(tuple_list):
    tuple_list_changed = []
    for letterpair in tuple_list:
        opp_xyz = transform_a_to_x[letterpair[1]] #XYZ:
        print(letterpair)
        print('opp choice', letterpair[1])
        
        if letterpair[0] == 'X': % should LO
            letter_changed = transform_a_to_x[wins_from_x[opp_xyz]] #you need to lose
            
        elif letterpair[0] == 'Y': % #shoud win
            letter_changed = transform_a_to_x[draw_from_x[opp_xyz]]
        elif letterpair[0] == 'Z':
            letter_changed = transform_a_to_x[lose_from_x[opp_xyz]] #you need to win
        else:
            print('fail',opp_xyz[0])
        print('your choice', letter_changed)
        tuple_list_changed.append((letter_changed,letterpair[1]))
    return tuple_list_changed
```

```python
tuple_list=preprocess_to_tuples(i = iE)
print(tuple_list[0:3])

# a_obj = a(i)
# a_obj.fun()

answA = scoring(tuple_list)
print(answA)
```

```python
def test_yield(init = 5):
    memory = init   
    add = 2
    while True:
        memory += add
        message = yield memory + add
        if message != None:
            add = message[0]
        else:
            add = 0

f_using_yield = test_yield(2.5)
f_using_yield.send(None)
print(next(f_using_yield))
f_using_yield.send([1, 0, 0])
print(next(f_using_yield))
f_using_yield.send([2, 0, 0])
print(next(f_using_yield))
f_using_yield.send([3, 0, 0])
print(next(f_using_yield))

```

```python
 def letter_generator(text):
     position = 0
     while True:
         message = yield text[position]
         if message!= None:
             position = message
         else:
             position += 1
 
letters = letter_generator("abcdefghijklmnopqrstucwxyz")
```
