```python
four_multipliers = [(1,1),(-1,-1),(-1,1),(1,-1)]

def parse(i=None):
    lines =i.split('\n')
    B_,S_=[],[]
    
    #print(len(lines))
    for line in lines:
        
        out=line.split(' ')
        #print(out)
        xpos = int(out[2][2:-1])
        ypos = int(out[3][2:-1])
        xbea = int(out[8][2:-1])
        ybea = int(out[9][2:])
        S_.append((xpos,ypos))
        B_.append((xbea,ybea))
        #print((xpos,ypos),',',(xbea,ybea))
    print('Parsed')
    return S_,B_ 


S_,B_ =parse(iE)
    
    
```

```python
# test and solve part A
# iA,iE,
#import scipy.sparse as sparse

class puzzle_grid:
    def __init__(self, i = iE,offset=(100,100),gridsize=(100,100), naive_method =False,y_crossing=20):
        self.gridsize = gridsize
        self.offset =offset
        self.S_, self.B_ = parse(i=i)
        self.y_crossing=y_crossing
        
        
        if naive_method:
            self.build_grid()        
            self.place_sensors()
            self.place_beacons()
            self.color_all_in_range()
            
            answA=None

        else:
            #calc_occ_on_one_line()
            min_max_on_line = self.calc_all_crossings(y=y_crossing)
            B_on_line_ = []
            S_on_line_ = []
            for x,y in self.S_:
                if y==  y_crossing:
                    S_on_line_.append(x)
                    
            for x,y in self.B_:
                if y== y_crossing:
                    B_on_line_.append(x)
            answA = count_nvals_in_intervals(min_max_on_line, S_on_line_, B_on_line_)
            print('answA',answA)
        self.answA = answA

        return
    def build_grid(self):
        
        
        N = self.gridsize[0] 
        M = self.gridsize[1] # 100+ self.offset[1]
        
        self.grid = np.zeros([N,M])
        #self.grid = sparse.csr_matrix([N,M])
    
            
    def place_beacons(self): # OBSOLETE
        for b in self.B_:
            self.grid[b[0], b[1]+self.offset[1]] = 5
            
        pass
            
    def place_sensors(self): # OBSOLETE
        for s in self.S_:
            self.grid[s[0]+self.offset[0] , s[1]+self.offset[1]] = 3
            
        pass
    def color_in_range(self,c=(0,0),dist = 10,value=8): # naive method => takes ages, ofcourse!

        for i in range(dist+1):
            for j in range(dist-i+1): #man dist 3 of minder.
            # if exactly manh dist => j = dist-i 
                for m in four_multipliers:
                    #nt('evaluated:',c[0]+i*m[0], c[1]+j*m[1])
                    #if c[0]+i*m[0]>=0 and c[1]+j*m[1]<=0:  # else it overflows! => to check if not out of range, not neg!
                    if not(self.grid[c[0]+i*m[0], c[1]+j*m[1]] >0):
                        self.grid[c[0]+i*m[0], c[1]+j*m[1]]=value 
        return 
    def color_all_in_range(self):
        for s,b in zip(self.S_, self.B_ ):
            self.color_in_range(c=(s[0]+self.offset[0],s[1]+self.offset[1]),dist = abs(s[0]-b[0])+abs(s[1]-b[1]),value=7)
            
    def calc_all_crossings(self, y=20):
        store_ =[]
        for s,b in zip(self.S_, self.B_):
            
            minmax=calc_crossing_min_max(s =s , y=y  ,dist = calc_manhattan(s,b))
            #print(s,b,minmax)
            if minmax is not  None:
                store_.append(minmax)
        return store_
            
def calc_manhattan(a=(0,0),b=(0,0)):
    return np.abs(a[0]-b[0]) +np.abs(a[1]-b[1]) #always pos

def calc_crossing_min_max(s =(10,10) , y=20 ,dist = 15):

    # WORK IN progress overlap1 =  dist - calc_manhattan(s,(s[0],y)) #int
    overlap =  calc_manhattan(s,(s[0],y))-dist +1
    
    
    
    
    #int #both alwasy posivite, point on circl
    #print('overlap ', overlap)
    if overlap==0 :
        return (s[0],s[0])
            
    else: #overlap >=0:
        #return (s[0]+overlap,s[0]-overlap)
        return (s[0]-overlap,s[0]+overlap)
                        
def count_nvals_in_intervals(tuplelist_from_to=[(10,20),(23,25),(23,35)], list_b_index_on_line = [],list_s_index_on_line= []): 
    counter = 0
    print(tuplelist_from_to)
    fullset_min_max =set()
    
        
    # make list.
    for (x,y) in tuplelist_from_to:
        fullset_min_max.add(x-1)
        fullset_min_max.add(x)
        fullset_min_max.add(x+1)
        fullset_min_max.add(y-1)
        fullset_min_max.add(y)
        fullset_min_max.add(y+1)
        #x & y are always ONE
    fullist = list(fullset_min_max)
    fullist.sort()
    print(fullist)

    #1,1,1 can be dropped.
    # (0,1,0) => 1

    count =0
    nprev = fullist[0]
    prev_In_sensorrange = False # left is outside.
    for n in fullist:
        In_sensorrange= False
        for (x,y) in tuplelist_from_to: # min max list!
            if not(In_sensorrange): # stop if safe.
                if (n >= x) and (n <=y):
                    #print(n, 'in ',(x,y))
                    In_sensorrange=True
        #print(n, safe_zone)
        '''if n in list_b_index_on_line: # or n in list_s_index_on_line:
                if not(In_sensorrange):
                    print(n, 'is beacon or sensor, on line, ignored')
                    pass # do nothing
                else:
                    count -= 1
                    print(n, 'is beacon or sensor, on line, count reduced ')
         '''   

        if not(prev_In_sensorrange) and In_sensorrange: # start counting from here, incl this one!
            prev_In_sensorrange = In_sensorrange
            nprev=n # Element left of it NOT in range, this one and next ones are
            
            
        if prev_In_sensorrange and not(In_sensorrange): # fron now on, not in range anymmore. count previous streak.
            print('count updated with', n - nprev)              
            count += abs(n - nprev)
            prev_In_sensorrange = In_sensorrange
            
            
        
        #print(n,safe_zone)
    #count += 1 #first is counted while it shoudnt
    
    return (count)
        
                
                
def fun2(i=None):
    print('world')
    pass
```
