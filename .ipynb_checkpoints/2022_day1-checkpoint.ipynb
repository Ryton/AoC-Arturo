```python
# run ONCE
puzzle = Puzzle(year=2022, day=1)
print(puzzle.user)
print(puzzle.input_data_fname)
print(puzzle.example_input_data_fname)

iA = puzzle.input_data
iE = puzzle.example_data
```

```python
# test and solve part A
# iA,iE

def count_max_calories(i):
    bin_per_elf = i.split("\n\n")
    curr_max = 0
    for e_list in bin_per_elf:
        counts = sum([int(el) for el in e_list.split("\n")])
        curr_max = max(curr_max, counts)
    return(curr_max)


answA = count_max_calories(iA[:])
```

```python
# test and solve part B, use inputs & test inputs.

def count_all_calories(i):
    bin_per_elf = i.split("\n\n")
    keep_all = []
    for e_list in bin_per_elf:
        counts = sum([int(el) for el in e_list.split("\n")])
        keep_all.append(counts)
        keep_all.sort(reverse=True)
    return sum(keep_all[0:3])
    pass


answB = count_all_calories(iA)
```
