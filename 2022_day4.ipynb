```python
# run ONCE 
puzzle = Puzzle(year=2022, day=4)
print(puzzle.user)
print(puzzle.input_data_fname)
print(puzzle.example_input_data_fname)

iA = puzzle.input_data
iE = puzzle.example_data
print(iA.split('\n')[0:3])
print(len(iA.split('\n')))
```

```python
# test and solve part A
#iA,iE, 

    
def make_sets(i=None):
    counter =0
    temp =0
    for line in i.split('\n'): # []'4-6,4-9', ...]
        list_edges = line.split(',') #  ['4-6','4-9']
        edges1 = list_edges[0].split('-')  #['4','6']
        edges2 = list_edges[1].split('-')
        if int(edges2[0])<=int(edges1[0]) and int(edges1[1])<=int(edges2[1]): #set 1 in set 2
            counter+=1
            if temp < 20:
                temp += 1
                #print('set1 in',line, edges1,edges2)
        elif int(edges1[0])<=int(edges2[0]) and int(edges2[1])<=int(edges1[1]): #set 1 in set 2
            counter+=1
            if temp < 20:
                temp += 1
                #print('set2 in', line, edges1,edges2)            
    return counter
    pass

print(make_sets(iA))
```

```python
# i = iE
# a_obj = a(i)
# a_obj.fun()

answA = make_sets(iA)
print(answA)
```

```python
puzzle.answer_a = answA # this submits!
```

```python
# test and solve part B, use inputs & test inputs.
def overlap_some(i=None):
    counter =0
    temp =0
    for line in i.split('\n'): # []'4-6,4-9', ...]
        list_edges = line.split(',') #  ['4-6','4-9']
        edges1 = list_edges[0].split('-')  #['4','6']
        edges2 = list_edges[1].split('-')
        list1= range(int(edges1[0]),int(edges1[1])+1)
        list2= range(int(edges2[0]),int(edges2[1])+1)
        if int(edges2[0]) in list1 or int(edges2[1]) in list1: #set 1 in set 2
            counter+=1
        elif int(edges1[0]) in list2 or int(edges1[1]) in list2: #set 1 in set 2
            counter+=1
            if temp < 20:
                temp += 1
                #print('set2 in', line, edges1,edges2)            
    return counter
    pass
print(overlap_some(iE))
```

```python
answB= overlap_some(i=iA)
print(answB)
puzzle.answer_b = answB # this submits!
```

```python
def test_yield(init = 5):
    memory = init   
    add = 2
    while True:
        memory += add
        message = yield memory + add
        if message != None:
            add = message[0]
        else:
            add = 0

f_using_yield = test_yield(2.5)
f_using_yield.send(None)
print(next(f_using_yield))
f_using_yield.send([1, 0, 0])
print(next(f_using_yield))
f_using_yield.send([2, 0, 0])
print(next(f_using_yield))
f_using_yield.send([3, 0, 0])
print(next(f_using_yield))

```

```python
 def letter_generator(text):
     position = 0
     while True:
         message = yield text[position]
         if message!= None:
             position = message
         else:
             position += 1
 
letters = letter_generator("abcdefghijklmnopqrstucwxyz")
```
