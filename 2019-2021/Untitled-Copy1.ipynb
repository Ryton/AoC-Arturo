```python
import pandas as pd
data = pd.read_csv('input_2.txt',header=None)
list = data.values
print('head',list[1:5],'\n tail:',list[-5:-1])
```

```python
forward,updown,X = [],[],[]
for i in list:
    if i[0][0]=='u':
        updown.append(-1)
        forward.append(0)
    elif i[0][0]=='d':
        updown.append(1)
        forward.append(0)
    elif i[0][0]=='f':
        updown.append(0)
        forward.append(1)
    X.append(int(i[0][-1]))
print(len(X),len(forward),len(updown))
```

```python
depth, hor, aim =0,0,0
for index in range(len(forward)):
    if forward[index]==0: #up or down.
        #depth += updown[index]*X[index]
        aim +=  updown[index]*X[index]
    else: # forward
        hor += abs(X[index])
        depth += aim * abs(X[index])
        
print (hor*depth)
        
```
