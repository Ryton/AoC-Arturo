```python
llist = [chr(i) for i in range(97,97+26)] # letters
val = [(i) for i in range(1,27)] # letters
l2v = dict(zip(llist,val))
v2l = dict(zip(val,llist))


go_mainly_left = False
if go_mainly_left:
    options=[(1+0.*1j),(0+1j),(0-1j),(-1+0.*1j)] # 4 directions in order of pref
else:
    options=[(-1),(1j),(-1j),(-1)] # 4 directions in order of pref


#print(dict_nextstep[(1+1j)])

def parse(i=None):
    grid = []
    rows = i.split('\n')
    nrows = len(rows)
    ncols = len(rows[0])
    print(nrows)
    print(ncols)
    for a in np.arange(0,nrows):
        for b in np.arange(0,ncols):
            grid.append(a+b*1j)
    #print(grid)
    
    letters = ''.join(i.split('\n'))
    startindex = letters.find('S')
    endindex = letters.find('E')

    letters=letters.replace('S','a')
    letters=letters.replace('E','z')
    #print(letters)   
    
    #g2v = l2g, lheights.values()
    
    endindex =grid[endindex]
    startindex =grid[startindex]
    
    return letters,grid ,startindex,endindex

# setup
# #what task
inputgrid = iE
[letters,grid,startindex,endindex]= parse(inputgrid)
intermediate_index = (20+(181-7)*1j)
intermediate_index = endindex

g2l = dict(zip(grid,letters))
l2g =dict(zip(letters,grid))

p2v = dict(zip(grid,[l2v[l]for l in letters]))
v2p =dict(zip([l2v[l]for l in letters],grid))
print(endindex)
print(startindex)
print(intermediate_index)
print('letter at intermed ',g2l[intermediate_index])
'''
print('p2v',p2v[(1+2*1j)])
print(p2v)
print(v2p)



#print(g2v)
# find a path around, clockwise of curr direction, take path
# if dead end, prune and backtrack).
print(v2p[26])
print(grid)
'''
v2p =dict(zip([l2v[l]for l in letters],grid))

# end of setup
```

```python
# test and solve part A
# iA,iE,
MAXITERATIONS=500000
explored_set = set()
iteration = 0
START = startindex
TARGET = endindex


def depth_first_shortest(shortest_paths,path, explored_set,iteration=0,success=False):
    currpos = path[-1]
    explored_set.add(currpos)
    
    iteration +=1
    #print(iteration)
    #print(currpos)
    if success or iteration>MAXITERATIONS:
        return shortest_paths,path, explored_set,iteration,success
    elif currpos == TARGET:
        success=True
        #print(startindex,'found')
        return shortest_paths,path,explored_set,iteration,success
    else:
        available_nodes= []
        #print(currpos)
        currval = p2v[currpos]
        #print(currval)
        print('options',options)
        for N in options:
            nextpos=N+currpos   # 4 options   
            print('check ', nextpos)
            if nextpos in grid:
                nextval=p2v[nextpos]
                #check if this is the shortest route
                curr_pathlength = len(path)
                stored_pathlength = len(shortest_paths[nextpos])
                if stored_pathlength==0:
                        path.append(nextpos)
                        shortest_paths[nextpos]= path
                        path.pop()
                        #if nextpos in explored_set:
                        #    explored_set.remove(nextpos) # => revisit that one!
                        print('added to shortlist')
                elif curr_pathlength+1 < stored_pathlength: # add length even if not visited.
                        path.append(nextpos)
                        shortest_paths[nextpos]= path
                        path.pop()
                        print('better route found ')
                
                if nextpos not in explored_set: # if not visited => revisit
                    print(nextpos,'added to available_nodes')
                    if abs(nextval - currval)<=1: # only up or equal for now
                        available_nodes.append(nextpos)
                        
                    print('available node:',available_nodes)
                    #elif nextval == currval:
                    #    available_nodes.append(nextpos)
                    ##dont ever go down again unless you really have to(end of the list)
                    #elif nextval == currval-1:
                    #    available_nodes.insert(0,nextpos)
                        
                else: # if already explored.
                    print(nextpos,'NOT valid for  available_nodes')
                    if len(path)+1 < len(shortest_paths[nextpos]):
                        path.append(nextpos)
                        shortest_paths[nextpos]= path
                        #path.pop()
                        print('here')
            else: # if already explored.
                print(nextpos,'outside grid')
    #setup!
    
    print('checkin now;', available_nodes)
    #then execute:
    for nextpos in available_nodes:
        if not success:
            path.append(nextpos)
            shortest_paths,path, explored_set,iteration,success = depth_first_shortest(shortest_paths,path, explored_set,iteration,success)
            # if unsuccessfull, this returns:
    return shortest_paths,path, explored_set,iteration,success


```
