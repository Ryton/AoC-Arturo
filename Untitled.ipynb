```python

from typing import List


def part1(lines: List[str]):
    char_count = 0
    symbol_count = 0
    for line in lines:
        char_count += len(line)
        subline = line[1:-1]
        it = iter(subline)
        result = ""
        try:
            while ch := next(it):
                if ch == "\\":
                    next_ch = next(it)
                    if next_ch != "x":
                        result += next_ch
                    else:
                        code = ''.join((next(it), next(it)))
                        result += "X"
                else:
                    result += ch
        except StopIteration:
            pass
        symbol_count += len(result)
    return char_count - symbol_count
part1(lines)
```
