```python
import numpy as np
import itertools
import ast
import operator as op
```

```python
FunNames = ['+','-','*','/','**']
def CalcRandcomb(source=123,target=6):
    output = []
    numbersStr= list(str(source))
    nOperatorsInFormula=len(str(source))-1
    proposedOperators = itertools.product(FunNames, repeat=nOperatorsInFormula)
    #print(len(proposedOperators) +"functions to evaluate.")
    for operators in proposedOperators:
        formula = ''.join([x for x in itertools.chain.from_iterable(itertools.zip_longest(numbersStr,operators)) if x])
        
        result = calcStr(formula)
        if type(result) is int:
            if result == target:
                output.append(formula) 
        #except: 
        #    pass
    return output

def custpow(a, b):
    # to fix long calc time.
    #if any(abs(n) > 3 for n in [a, b]):
    if b > 3 :
        raise ValueError((a,b))
    return op.pow(a, b)

def custmult(a, b):
    # to fix long calc time.
    if any(abs(n) > 1E5 for n in [a, b]):
        raise ValueError((a,b))
    return op.mul(a, b)


operators = {ast.Add: op.add, ast.Sub: op.sub, ast.Mult: custmult,
             ast.Div: op.truediv, ast.Pow: custpow, ast.BitXor: op.xor,
             ast.USub: op.neg}

def calcStr(formula):
    return eval(formula)
    # avoiding evil-EVAL:
    # https://stackoverflow.com/questions/2371436/evaluating-a-mathematical-expression-in-a-string
    return eval_(ast.parse(formula, mode='eval').body)

def eval_(node):
    if isinstance(node, ast.Num): # <number>
        return int(node.n)
    elif isinstance(node, ast.BinOp): # <left> <operator> <right>
        return int(operators[type(node.op)](eval_(node.left), eval_(node.right)))

    elif isinstance(node, ast.UnaryOp): # <operator> <operand> e.g., -1
        return int(operators[type(node.op)](eval_(node.operand)))
    else:
        return 0#raise TypeError(node)
```

```python
print(CalcRandcomb(1445,21))
print(CalcRandcomb(321,6))
print(CalcRandcomb(10324,10))
```
