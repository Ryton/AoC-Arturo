```python
# test and solve part A
# iA,iE,
def buildriskmap(i=None):
    i_lines = i.split('\n')
    #print(i_lines)
    nrows = len(i_lines)
    ncols = len(i_lines[0])
    print(nrows,ncols)
    
    riskmap = np.zeros([nrows,ncols])
    for j in range(nrows):
        for k in range(ncols):
            
            letters = i_lines[k]
            riskmap[j,k] =letters[j]
    return riskmap

def extend_riskmap(riskmap=None,extension =5):
    
    B_riskmap = np.zeros([riskmap.shape[0]*extension,riskmap.shape[1]*extension])
    nrows = riskmap.shape[1]
    ncols = riskmap.shape[0]
    for a in range(extension):
        for b in range(extension):
            for j in range(nrows):
                for k in range(ncols):
            
                    B_riskmap[nrows*a+j,ncols*b+k] = np.mod(a+b + riskmap[j,k]-1,9)+1
                          
    #print(i_lines)
    
    return B_riskmap

def build_start_costmap(riskmap):
    nrows = riskmap.shape[0]
    ncols = riskmap.shape[1]
    costmap = np.ones([nrows,ncols])*99999
    for j in range(0,nrows): # fill first row, starting from row 1!
        if j ==0:
            costmap[j,0]= 0
        else:
            costmap[j,0]= riskmap[j,0] +costmap[j-1,0]
        
    for k in range(0,ncols):
        if k ==0:
            costmap[0,k]= 0
        else:
            costmap[0,k]= riskmap[0,k] +costmap[0,k-1]
    return costmap

    

def buildcost_map(riskmap=None,n_iter_max = 85):
    nrows = riskmap.shape[1]
    ncols = riskmap.shape[0]
    costmap = np.ones([nrows,ncols])*99999
    costmap_prev = np.ones([nrows,ncols])*99999
    costmap_t_min2 = np.ones([nrows,ncols])*-9999
    flipflop = True
    niter = 0
    costmap= build_start_costmap(riskmap)
    
    while (niter< n_iter_max): # and not((costmap == costmap_t_min2).all()):
        costmap_t_min2 = costmap_prev
        costmap_prev =costmap
        #print(niter)

        for j in range(0,nrows): # check ROWS, starting from row 1!
            for k in range(0,ncols):
                #min(costmap[j,k],min(costmap[j-1,k]+costmap[j,k]) 
                curr = costmap[j,k]
                if (0<j < nrows-1):
                    leftright = min(costmap[j+1,k]+riskmap[j,k],costmap[j-1,k]+riskmap[j,k])                                
                elif j==0:
                    leftright = costmap[j+1,k]+riskmap[j,k]
                elif j == nrows-1:
                    leftright = costmap[j-1,k]+riskmap[j,k]
                    
                if curr > leftright:
                    costmap[j,k] = leftright
                    

                if (0<k<ncols-1):
                    #print(k)
                    topbottom = min(costmap[j,k-1]+riskmap[j,k],costmap[j,k+1]+riskmap[j,k])
                elif k==0:
                    topbottom = costmap[j,k+1]+riskmap[j,k]
                elif k == ncols-1:
                    topbottom = costmap[j,k-1]+riskmap[j,k]

                if curr > topbottom:
                    costmap[j,k] = topbottom

        niter += 1
        print(niter)
    return costmap
```

```python
#task A
riskmap=buildriskmap(iE)
print(riskmap)
start_costmap = build_start_costmap(riskmap)
#print(start_costmap)
costmap=buildcost_map(riskmap)
answA = costmap[-1,-1]
print(costmap)
print(answA)
```
