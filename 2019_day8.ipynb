```python
def build_picture(layerlist,nlayers=100,npixels=25*6):
    #0 is black, 1 is white, and 2 is transparent.
    fin_picture=[]
    for i in range(npixels):
        for j in range(nlayers):
            if layerlist[j][i]=='1':
                fin_picture.append(1) 
                break
            if layerlist[j][i]=='0':
                fin_picture.append(0) 
                break
    return fin_picture
fin=build_picture(layerlist)
print(fin)
```

```python
def test_yield(init = 5):
    memory = init   
    add = 2
    while True:
        memory += add
        message = yield memory + add
        if message != None:
            add = message[0]
        else:
            add = 0

f_using_yield = test_yield(2.5)
f_using_yield.send(None)
print(next(f_using_yield))
f_using_yield.send([1, 0, 0])
print(next(f_using_yield))
f_using_yield.send([2, 0, 0])
print(next(f_using_yield))
f_using_yield.send([3, 0, 0])
print(next(f_using_yield))

```

```python
 def letter_generator(text):
     position = 0
     while True:
         message = yield text[position]
         if message!= None:
             position = message
         else:
             position += 1
 
letters = letter_generator("abcdefghijklmnopqrstucwxyz")
```
